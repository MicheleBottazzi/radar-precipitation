```python
import xarray as xr 
import numpy as np
import regionmask
import geopandas as gpd
import pandas as pd
import matplotlib.pyplot as plt
import warnings; warnings.filterwarnings(action='ignore')
%matplotlib inline
```

```python
PATH_TO_SHAPEFILE = 'Italy_boundaries.shp'
countries = gpd.read_file(PATH_TO_SHAPEFILE)
countries.head()
```

```python
my_list = list(countries['CNTR_ID'])
my_list_unique = set(list(countries['CNTR_ID']))
indexes = [my_list.index(x) for x in my_list_unique]
```

```python
my_list
```

```python
model=''
t_in = ''        
data = 'r-factor_r1_regrid_v2_seconds.nc'
ds = xr.open_dataset(data)
ds
```

```python
ds = ds.assign_coords(longitude=(((ds.lon + 180) % 360) - 180)).sortby('lon')
ds
```

```python
len(list(countries.geometry.values[i] for i in indexes))
len(list(countries.CNTR_ID[indexes]))
len(indexes)
```

```python
a = range(0,countries.shape[0])
np.shape(a)
```

```python
countries_mask_poly = regionmask.Regions(
    name = 'CNTR_ID', 
    numbers = indexes, 
    names = countries.CNTR_ID[indexes], 
    abbrevs = countries.CNTR_ID[indexes], 
    outlines = list(
        countries.geometry.values[i] for i in range(0,countries.shape[0])
    ))
countries_mask_poly
```

```python
ds
```

```python
mask = countries_mask_poly.mask(ds, lat_name='lat', lon_name='lon')
mask
```

```python
mask.to_netcdf('/home/michele/radar-precipitation/clipped.nc') 
```

```python
mask.max()
```

```python
plt.figure(figsize=(16,8))
ax = plt.axes()
mask.plot(ax = ax)
countries.plot(ax = ax, alpha = 0.8, facecolor = 'none', lw = 1)
```

# ########################################

```python
ID_COUNTRY = 0
print(countries.CNTR_ID[ID_COUNTRY])
```

```python
lat = mask.lat.values
lon = mask.lon.values

sel_mask = mask.where(mask == ID_COUNTRY).values
sel_mask.shape
```

```python
id_lon = lon[np.where(~np.all(np.isnan(sel_mask), axis=0))]
id_lat = lat[np.where(~np.all(np.isnan(sel_mask), axis=1))]
id_lat.shape
```

```python
np.isnan(sel_mask)
```

```python
out_sel1 = ds.sel(
    lat = slice(id_lat[0], id_lat[-1]), 
    lon = slice(id_lon[0], id_lon[-1])).compute().where(mask == ID_COUNTRY)
```

```python
plt.figure(figsize=(12,8))
ax = plt.axes()
out_sel1.rf.plot(ax = ax)
countries.plot(ax = ax, alpha = 0.8, facecolor = 'none')
```
